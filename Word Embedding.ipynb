```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
text  = ['the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
from keras.preprocessing.text import one_hot
```

```python
vocab_size = 1000
```

```python
one_hot_vector = [one_hot(words,vocab_size) for words in text]
```

```python
print(one_hot_vector)
```

### Word Embedding Representation

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
max_legnth = 6
embedded_text = pad_sequences(one_hot_vector,padding = 'pre',maxlen=max_legnth)
```

```python
print(embedded_text)
```

```python
dim = 10
```

```python
## Creating Model 
```

```python
model = Sequential()
model.add(Embedding(vocab_size,10,input_length = max_legnth))
model.compile("adam","mse")
```

```python
model.summary()
```

```python
print(model.predict(embedded_text))
```
