```python
import sklearn
import numpy as np
import pandas as pd
```

```python
import nltk 
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
from nltk.corpus import stopwords
```

```python
vector_corpus = ['ذهب احمد الي المول ليشتري ملابس جديده','سافر احمد الي الولايات المتحدة الامريكية ليكمل دراساته','استقبل محمود احمد في المطار']
```

```python
arabic_stop_words = stopwords.words("Arabic")
```

```python
print(arabic_stop_words)
```

```python
vectorizer = TfidfVectorizer(stop_words = arabic_stop_words)
```

```python
tfidf = vectorizer.fit_transform(vector_corpus)
```

```python
print(vectorizer.get_feature_names())
```

```python
tfidf.shape
```

```python
df_tfidf = pd.DataFrame(tfidf.toarray())
```

```python
df_tfidf.head()
```

```python
column_names = vectorizer.get_feature_names()
```
